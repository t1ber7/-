```python
lesson = []
```

```python
for i in df1.variable:
    lesson.append(i.split("_"))
```

```python
lst = []
```

```python
for i in lesson:
    new_value = i[0].replace("РегиональныйЭтап", "")
    i[0] = new_value.replace("(", "").replace(")", "")
    lst.append(i)
```

```python
state = []
subject = []
date = []
```

```python
lst
```

```python
for i in lst:
    state.append(i[0])
    subject.append(i[1])
    date.append(i[2])
```
