```python
df = pd.read_csv('passengers.csv', index_col = 'Month', parse_dates = True)
```

```python
df['1949-08':'1950-03']
```

```python
df.plot();
```

```python
ax = df.plot(figsize = (12,6), legend = None)
ax.set(title = 'Перевозки пассажиров с 1949 по 1960 год', xlabel = 'Года', ylabel = 'Количество пассажиров');
```

```python
plt.figure(figsize = (15,8))
 
plt.plot(df, label = 'Перевозки пассажиров по месяцам', color = 'steelblue')
plt.plot(df.rolling(window = 12).mean(), label = 'Скользящее среднее за 12 месяцев', color = 'orange')
 
plt.legend(title = '', loc = 'upper left', fontsize = 14)
 
plt.xlabel('Месяцы', fontsize = 14)
plt.ylabel('Количество пассажиров', fontsize = 14)
plt.title('Перевозки пассажиров с 1949 по 1960 год', fontsize = 16)
 
plt.show()
```
