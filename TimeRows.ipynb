```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('passengers.csv', index_col = 'Month', parse_dates = True)
```

```python
df['1949-08':'1950-03']
```

```python
df.plot();
```

```python
ax = df.plot(figsize = (12,6), legend = None)
ax.set(title = '   1949  1960 ', xlabel = '', ylabel = ' ');
```

```python
plt.figure(figsize = (15,8))
 
plt.plot(df, label = '   ', color = 'steelblue')
plt.plot(df.rolling(window = 12).mean(), label = '   12 ', color = 'orange')
 
plt.legend(title = '', loc = 'upper left', fontsize = 14)
 
plt.xlabel('', fontsize = 14)
plt.ylabel(' ', fontsize = 14)
plt.title('   1949  1960 ', fontsize = 16)
 
plt.show()
```
