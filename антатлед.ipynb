# Пояснение данных
* ### country - Страна
* ### region - регион
* ### happiness_score - Уровень счастья
* ### gdp_per_capita - ВВП страны на человека
* ### social_support - социальная поддержка
* ### healthy_life_expectancy - забота о здоровом образе жизни
* ### freedom_to_make_life_choices - свобода выбора
* ### generosity - Щедрость
* ### perceptions_of_corruption - восприятие коррупции

```python
forheat = df[["happiness_score", "gdp_per_capita", "social_support", "healthy_life_expectancy", "freedom_to_make_life_choices", "generosity", "perceptions_of_corruption"]].corr()
forheat
```

```python
sns.heatmap(data=forheat, annot=True);
```

### Зависимости
* мы видим зависимости уровни счастья на ВВП на человека
* так же видим что щедрость людей мало сказывается на уровне счастья
* Сильно влияют 3 показателя на уровень счастья, это: ВВП на человека, социальная помощь, отношение к ЗОЖ, Свобода выбора.

### Видим что отношение к корупции у людей негативное

```python
plt.figure(figsize=(10, 5))
plt.tight_layout()

sns.histplot(df["gdp_per_capita"])
plt.title(f'ВВП страны')
plt.xlabel("gdp_per_capita")
plt.ylabel('Frequency')
plt.show();
```

* Мы видим что записи ВВП встречаются около значение 1, примерно 30 раз

```python
plt.figure(figsize=(10, 8))
plt.tight_layout()

sns.histplot(df[["social_support", "happiness_score"]])
plt.title(f'Значения соц поддержки для уровня счастья')
plt.xlabel("social support")
plt.ylabel('Frequency')
plt.show();
```

* Мы можем наблюдать что значение соц помощи держится примерно в значении 1, однако уровень счастья очень разница, несмотря на почти равное значение соц помощи.
